```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
from tqdm import tqdm
```

```python
li_main = [i for i in range(0,67600)]
```

```python
li_main.insert(0,'label')
```

```python
f=open('AD-MCI-CN.csv','a')
writer = csv.writer(f)
writer.writerow(li_main)

for folder in os.listdir("../../../data/train"):
    for items in tqdm(os.listdir("../../../data/train/" + folder)):
        li = []
        if folder == 'AD' :
            li.append(0)
#             break
        elif folder == 'MCI' :
            li.append(1)
        elif folder == 'CN' :
            li.append(2)
        img = plt.imread("../../../data/train/" + folder + '/' + items)       
        li.extend(list(img.flatten()))
        writer.writerow(li)
        
f.flush()
f.close()        
```
